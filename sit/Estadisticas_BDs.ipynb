# DETECCION DE OBJETOS

Importamos las librerias necesarias

```python
import os
import python_voc_parser as voc 
import pandas as pd
from IPython.display import display, HTML
```

Definimos funciones auxiliares

```python
def detection_db_stats(database_directory, set_filename):  
    IMAGE_SET_PATH = os.path.join(database_directory, 'ImageSets','Main', set_filename)
    IMAGE_PATH = os.path.join(database_directory, 'JPEGImages')
    ANNOTATIONS_PATH = os.path.join(database_directory, 'Annotations')

    # Creamos un parser 
    parser = voc.VocAnnotationsParser(IMAGE_PATH, IMAGE_SET_PATH, ANNOTATIONS_PATH)

    # Para obtener una lista con todos los datos
    # annotations_list = parser.annotation_line_list

    # Para obtener un dataframe
    df = parser.get_annotation_dataframe()
    
    print('Ejemplo de datos:')
    display(df.head())
    
    classes = pd.DataFrame(df.class_name.value_counts().reset_index())
    classes.columns = ['class_name', 'count']

    num_images = sum(1 for line in open(IMAGE_SET_PATH) if line.rstrip())
    num_classes = classes.shape[0]

    print("")
    print("")
    print('Estadísticas Base de Datos_________________')
    print('Numero de Imagenes: {}'.format(num_images))
    print('Numero de clases  : {}'.format(num_classes))
    print('Objetos por clase : ')
    display(classes)
    #display(HTML('<h3>Hello, world!</h3>'))
    #print("Here's a link:")
    #display(HTML("<a href='http://www.google.com' target='_blank'>www.google.com</a>"))
    #print("some more printed text ...")
    #display(HTML('<p>Paragraph text here ...</p>'))
```

# CLASIFICACIÓN

```python
def classification_db_stats(dir_db):    
    dir_root = os.path.join(dir_db, 'images')
    directories = []
    class_names = []
    class_counts = []
    num_images = 0
    for filename in os.listdir(dir_root):
        path = os.path.join(dir_root, filename)
        if os.path.isdir(path):
            class_names.append(filename)
            count = 0
            for filename in os.listdir(path):
                count += 1
            class_counts.append(count)
            num_images += count

    df = pd.DataFrame({ 'class_name':class_names, 'count':class_counts  })

    print('Estadísticas Base de Datos_________________')
    print('Path              : {}'.format(dir_db))
    print('Numero de Imagenes: {}'.format(num_images))
    print('Numero de clases  : {}'.format(len(class_names)))
    print('Objetos por clase : ')
    display(df.sort_values('count', ascending=False ))
    #df.style.set_table_styles([
    #    {'selector': '.row_heading, .blank', 'props': [('display', 'none;')]}
    #])

```

## Trucks

```python
DATABASE_DIRECTORY = '/root/databases/Trucks'
classification_db_stats(DATABASE_DIRECTORY)
```
